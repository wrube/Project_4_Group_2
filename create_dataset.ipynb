```python
# List of columns to drop, considering possible duplication
columns_to_drop = ['Card Number', 'Expires', 'CVV', 'Credit limit', 'Acct Open Date']

# Strip any extra spaces in the column names to ensure proper identification
cards_df.columns = cards_df.columns.str.strip()

# Drop the specified columns
filtered_cards_df = cards_df.drop(columns=[col for col in columns_to_drop if col in cards_df.columns])

# Display the DataFrame to verify the changes
filtered_cards_df.head()
```

```python
# List of columns to drop
columns_to_drop = [
    'Year', 'Amount', 'Merchant Name', 'Merchant State', 'Zip'
]

# Strip any extra spaces in the column names to ensure proper identification
filtered_transactions_500users.columns = filtered_transactions_500users.columns.str.strip()

# Drop the specified columns
filtered_transactions_df = filtered_transactions_500users.drop(columns=[col for col in columns_to_drop if col in transactions_df.columns])

# Display the DataFrame to verify the changes
filtered_transactions_df.head()
```

```python
# First, merge filtered_transactions_df with filtered_users_df on 'User'
merged_df = pd.merge(filtered_transactions_df, filtered_users_df, on='User', how='inner')

# Then, merge the resulting DataFrame with filtered_cards_df on 'User'
# final_merged_df = pd.merge(merged_df, filtered_cards_df, on='User', how='inner')
final_merged_df = merged_df.merge(cards_df, how='inner', left_on=['User', 'Card'], right_on=['User', 'CARD INDEX'])

# Set the option to display all columns
pd.set_option('display.max_columns', None)

# Display the final merged DataFrame to verify the changes
final_merged_df.head()
```
